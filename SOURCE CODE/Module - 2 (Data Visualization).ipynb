# Data Visualization

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Maternal_Health.csv')
```

```python
df=data
df.columns
```

```python
# Histogram
df['Age'].plot.hist()
plt.xlabel('AGE')
plt.ylabel('TOTAL MEMBERS')
plt.title('AGE DISTIBUTION IN PREGNANT WOMENS')
```

```python
# Histogram
df['SystolicBP'].plot.hist()
plt.ylabel('Counts')
plt.xlabel('SystolicBP')
```

```python
# Histogram
df['DiastolicBP'].plot.hist()
plt.ylabel('Counts')
plt.xlabel('DiastolicBP')
```

```python
# Histogram
df['HeartRate'].plot.hist()
plt.ylabel('Counts')
plt.xlabel('Heart Rate')
```

```python
plt.figure(figsize=(8,4))
plt.hist(df['BS'])
plt.title('Average Blood Sugar Levels')
plt.xlabel('Blood Sugar Level')
plt.ylabel('Total Patients')
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.stripplot(y=df['RiskLevel'],x=df['Age'],ax=ax)
plt.title('Age Wise Risk Levels')
plt.show()
```

```python
def RiskLevel(df,variable):
    dataframe_pie=df[variable].value_counts()
    ax=dataframe_pie.plot.pie(figsize=(9,9),autopct='%1.2f%%',fontsize=15)
    ax.set_title(variable+' \n',fontsize=10)
    return np.round(dataframe_pie/df.shape[0]*100,2)
RiskLevel(df, 'RiskLevel')
```
